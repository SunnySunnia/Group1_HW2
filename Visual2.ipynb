```python
import urllib
from pandas import read_csv

url= 'http://earthquake.usgs.gov/earthquakes/catalogs/eqs7day-M1.txt'
data= read_csv(urllib.urlopen(url))
#dont need urllib. if said from urllib import urlopen

data[0:5]

clean_data = data.dropna(axis=0,how='any')
#pandas.Data Frame.dropna; axis:0-row, 1-col; how:want to drop the row that has any NA's;
#naming differently, better refering to specific dataframe, python allows running codes in any order any numbers of time;
#this 'clean_data' is only changing 'data';
#so no matter how many times you run this code, as long as 'data' is unchanged, 'clean_data' will always be the same.
clean_data[0:3]
alaska = clean_data[clean_data.Src == 'ak']
#clean_data.Src=='ak' tells which rows in the Src column have value =='ak' 
#clean_data[clean_data.Src=='ak'] :subsetting those rows
alaska[0:10]
```

```python
from mpl_toolkits.basemap import Basemap

lllon=-180
lllat=50
urlon=-120
urlat=72
state=alaska

def G1_quakes(state,lllon,lllat,urlon,urlat):
    m = Basemap(llcrnrlon=lllon, llcrnrlat=lllat,
                urcrnrlon=urlon,urcrnrlat=urlat,
                resolution='l', area_thresh=1000.,projection='merc')
    m.drawcoastlines()
    m.drawcountries()
    m.fillcontinents(color='coral', lake_color='aqua')
    m.drawmapboundary(fill_color='aqua')
    
    lon = state.Lon  
    lat = state.Lat
    xpt,ypt = m(lon,lat)
    lonpt, latpt = m(xpt,ypt,inverse=True)
    sizes=state.Magnitude
    for i in range(1,len(lon)):
        m.plot(xpt[i-1:i],ypt[i-1:i],'ro',markersize=sizes[i-1:i]*4,alpha=0.6)
    return m

m1=G1_quakes(state,lllon,lllat,urlon,urlat)
```
